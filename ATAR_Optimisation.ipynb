### 1-a

En tant que directeur financier des usines A et B, notre objectif est d'assurer un bénéfice maximum. Pour cela, nous devons optimiser la production et la distribution des composants mémoires de manière à maximiser les profits tout en respectant les contraintes de capacité de production et de demande.

La fonction à optimiser serait donc la fonction de profit total, qui est donnée par :

(prix de vente unitaire - coût de production unitaire) * (x_A1 + x_A2 + x_A3 + x_B1 + x_B2 + x_B3) - (1000* | x_A1 + x_B1 - 150 | + 1000* | x_A2 + x_B2 - 120 | + 1000* | x_A3 + x_B3 - 90 | )

Les variables de décisions restent les mêmes que dans l'énoncé précédent, x_ij où i est l'usine et j est le client.

Les contraintes restent les mêmes que dans l'énoncé précédent avec l'ajout de la contrainte :

Les clients ne peuvent pas obtenir la totalité de leurs commandes avec deux usines : x_A1, x_A2, x_A3, x_B1, x_B2, x_B3 >= 0

Nous pouvons utiliser la loi uniforme pour générer des valeurs aléatoires pour les coûts de livraison dans les intervalles donnés et utiliser ces valeurs pour résoudre notre modèle d'optimisation. Nous pouvons ensuite calculer les résultats pour chaque scénario et les comparer pour obtenir une idée des incertitudes liées aux coûts de livraison.

Pour mettre en oeuvre cette méthode, nous pouvons utiliser un script pour générer des valeurs aléatoires pour les coûts de livraison selon la loi uniforme et résoudre le modèle pour chaque scénario. Il est important de noter que cette méthode n'est pas une solution définitive mais elle permet de prendre en compte les incertitudes liées aux coûts de livraison et de fournir des informations sur les risques potentiels liés à ces incertitudes.
```python
### la solution de 1-d
import pulp
import numpy as np

# Define the decision variables
x1 = pulp.LpVariable('X1', lowBound=0, cat='Integer')
x2 = pulp.LpVariable('X2', lowBound=0, cat='Integer')
x3 = pulp.LpVariable('X3', lowBound=0, cat='Integer')
y1 = pulp.LpVariable('Y1', lowBound=0, cat='Integer')
y2 = pulp.LpVariable('Y2', lowBound=0, cat='Integer')
y3 = pulp.LpVariable('Y3', lowBound=0, cat='Integer')

# Define the production and delivery capacities
capacity_a = 160
capacity_b = 200

# Define the demands of the customers
demand_f1 = 150
demand_f2 = 120
demand_f3 = 90

# Define the intervals of variation for the delivery costs
variation_a = 0.25
variation_b = 0.35

# Generate a set of scenarios for the delivery costs
num_scenarios = 100
delivery_costs_a = np.random.uniform(low=1-variation_a, high=1+variation_a, size=(num_scenarios, 3))
delivery_costs_b = np.random.uniform(low=1-variation_b, high=1+variation_b, size=(num_scenarios, 3))

# Initialize lists to store the optimal solutions
optimal_costs = []
optimal_x1 = []
optimal_x2 = []
optimal_x3 = []
optimal_y1 = []
optimal_y2 = []
optimal_y3 = []

# Solve the linear programming problem for each scenario
for i in range(num_scenarios):
    # Create the linear programming problem
    lp_problem = pulp.LpProblem('Cost Minimization', pulp.LpMinimize)

    # Define the objective function
    lp_problem += delivery_costs_a[i, 0]*x1 + delivery_costs_a[i, 1]*x2 + delivery_costs_a[i, 2]*x3 + delivery_costs_b[i, 0]*y1 + delivery_costs_b[i, 1]*y2 + delivery_costs_b[i, 2]*y3

    # Define the constraints
    lp_problem += x1 + x2 + x3 <= capacity_a
    lp_problem += y1 + y2 + y3 <= capacity_b
    lp_problem += x1 + y1 == demand_f1
    lp_problem += x2 + y2 == demand_f2
    lp_problem += x3 + y3 == demand_f3
    lp_problem += x1 >= 0
    lp_problem += x2 >= 0
    lp_problem += x3 >= 0
    lp_problem += y1 >= 0
    lp_problem += y2 >= 0
    lp_problem += y3 >= 0

    # Solve the linear programming problem
    lp_problem.solve()

    # Store the optimal solutions
optimal_costs.append(pulp.value(lp_problem.objective))
optimal_x1.append(x1.varValue)
optimal_x2.append(x2.varValue)
optimal_x3.append(x3.varValue)
optimal_y1.append(y1.varValue)
optimal_y2.append(y2.varValue)
optimal_y3.append(y3.varValue)

# Compute the mean and standard deviation of the optimal costs
mean_cost = np.mean(optimal_costs)
std_cost = np.std(optimal_costs)

# Print the results
print(f'Mean optimal cost: {mean_cost:.2f}')
print(f'Standard deviation of optimal cost: {std_cost:.2f}')

```

### 2-a

Pour résoudre ce problème, nous pouvons utiliser la méthode de la programmation linéaire en nombre entiers (IP). Il s'agit d'une technique d'optimisation qui permet de trouver la solution optimale pour des problèmes où certaines variables doivent prendre des valeurs entières.

La méthode consiste à :

Formuler le problème sous forme de modèle mathématique, en définissant les variables de décisions, les contraintes et la fonction objectif.
Utiliser un solveur de programmation linéaire en nombre entiers pour résoudre le modèle.
Il existe plusieurs algorithmes pour résoudre les problèmes de programmation linéaire en nombre entiers tels que :

L'algorithme branch-and-bound qui consiste à créer une arborescence de recherche qui représente toutes les solutions potentielles du problème en utilisant des techniques de branchement et de bornage pour éliminer les solutions inférieures. Il permet de trouver la solution optimale en explorant tous les noeuds de l'arborescence.
L'algorithme branch-and-cut qui consiste à combiner l'algorithme branch-and-bound avec des techniques de résolution de programmation linéaire pour résoudre efficacement des problèmes de programmation linéaire en nombre entiers. Il utilise des coupures pour éliminer les solutions inférieures et utilise la résolution de programmation linéaire pour résoudre les sous-problèmes.

L'algorithme de colonne-et-ligne qui consiste à ajouter une colonne (variable) ou une ligne (contrainte) à chaque itération pour améliorer la solution courante jusqu'à ce qu'une solution optimale soit trouvée.
L'algorithme de recherche de voisinage qui consiste à explorer les solutions proches de la solution courante pour trouver une solution optimale.
Il est important de noter que la résolution d'un problème de programmation linéaire en nombre entiers peut être difficile et prendre beaucoup de temps, en particulier pour des modèles de grande taille et complexité. Il est donc important de choisir les algorithmes appropriés en fonction des caractéristiques de chaque problème spécifique.

```python
def discomfort_index(technician, task):
    preferences = {
        "T1": ["C1", "C2", "C4", "C5", "C6"],
        "T2": ["C2", "C4", "C6", "C5", "C3"],
        "T3": ["C1", "C3", "C6", "C2", "C4", "C5"],
        "T4": ["C1", "C4", "C5", "C3", "C2"],
        "T5": ["C2", "C3", "C1", "C5", "C4", "C6"]
    }
    return -1*(preferences[technician].index(task))

def discomfort_score(assignments):
    score = 0
    for technician, task in assignments.items():
        if x[technician, task].value() == 1:
            score += discomfort_index(technician, task)
    return score



```

L'algorithme de Hungarian est un algorithme de résolution optimale qui utilise une méthode de programmation linéaire pour trouver l'affectation optimale des tâches aux agents. Il est efficace pour résoudre des problèmes d'affectation à coûts fixes. Il est efficace pour les problèmes de petite à moyenne taille mais peut devenir très lent pour les problèmes de grande taille.

Le recuit simulé, d'autre part, est un algorithme de recherche heuristique qui utilise des méthodes de simulation pour trouver une solution approximative au problème d'affectation. Il est souvent utilisé pour résoudre des problèmes d'affectation à coûts variables et pour les problèmes de grande taille. Il peut générer des solutions de qualité inférieure à celle de l'algorithme de Hungarian, mais il est généralement plus rapide.

En résumé, l'algorithme de Hungarian est une solution optimale pour les problèmes d'affectation à coûts fixes, mais il peut devenir lent pour les problèmes de grande taille. Le recuit simulé est une solution approximative pour les problèmes d'affectation à coûts variables et les problèmes de grande taille, mais il peut générer des solutions de qualité inférieure. Il est donc important de choisir la meilleure méthode en fonction des spécificités du problème.

```python
from pulp import *

# Define the problem
prob = LpProblem("Technician Scheduling", LpMinimize)

# Technician availability constraints
technicians = ["T1", "T2", "T3", "T4", "T5"]
shifts = range(5)
tasks = ["A", "B", "C", "D", "E"]

# Define the decision variables
x = LpVariable.dicts("x", [(technician, shift, task) for technician in technicians for shift in shifts for task in tasks],
                      0, 1, LpBinary)

# Technician availability constraints
for technician in technicians:
    for shift in range(5):
        task_assignments = [x[technician, shift, task] for task in tasks]
        prob += lpSum(task_assignments) <= 1

# Task demand constraints
task_demand = {"A": 3, "B": 2, "C": 2, "D": 1, "E": 1}
for task in tasks:
    demand = task_demand[task]
    task_assignments = [x[technician, shift, task] for technician in technicians for shift in shifts]
    prob += lpSum(task_assignments) == demand

# Objective function
discomfort_scores = {technician: sum(x[technician, shift, task] * (5 - shift - tasks.index(task))
                                    for shift in range(5) for task in tasks)
                     for technician in technicians}
prob += lpSum([discomfort_scores[technician] for technician in discomfort_scores])

# Solve the problem
prob.solve()

# Print the solution
for v in prob.variables():
    if v.varValue:
        print(v.name, "=", v.varValue)
print("Total discomfort = ", value(prob.objective))




```

```python
import gurobipy as gp

# Données du problème
nb_techniciens = 5
nb_taches = 6
preference = {
    0: [0, 1, 3, 4, 5],
    1: [1, 3, 5, 2],
    2: [0, 2, 5, 3, 4],
    3: [0, 3, 2, 1],
    4: [1, 2, 0, 4, 5]
}
demande = [3, 4, 3, 5, 5, 5]

# Création du modèle
m = gp.Model()

# Variables de décision
x = m.addVars(nb_techniciens, nb_taches, vtype=gp.GRB.BINARY, name="x")

# Contraintes
# La somme des tâches effectuées par chaque technicien doit être égale à 1
m.addConstrs((x.sum(i, '*') == 1 for i in range(nb_techniciens)), name="c1")

# La somme des tâches effectuées par chaque technicien pour chaque créneau horaire doit être inférieure ou égale à la demande
m.addConstrs((x.sum('*', j) <= demande[j] for j in range(nb_taches)), name="c2")

# Fonction objectif
inconfort = {(i, j): -k for i in range(nb_techniciens) for j, k in enumerate(preference[i])}
m.setObjective(x.prod(inconfort), gp.GRB.MINIMIZE)

# Optimisation
m.optimize()

# Affichage des résultats
print("La fonction objectif optimale est : ", abs(m.objVal))
for i in range(nb_techniciens):
    for j in range(nb_taches):
        if x[i, j].x > 0:
            print("Le technicien {} effectue la tâche {}".format(i, j))
```
